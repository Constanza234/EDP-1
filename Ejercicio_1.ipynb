```python
#Librerias
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
```

Funciòn a trozos

```python
def f1(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 0
    return y
def f2(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 2- x[i]
    return y
```

```python
x1 = np.linspace(-2,0,1000)  # linspace (valor-inicial, valor-final, número de valores)
x2 = np.linspace(0,2,1000)
y1 = f1(x1)
y2 = f2(x2)
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=r'$f(x) = x + 2$ para $-2 \leq x \leq 0$')
plt.plot(x2, y2, label=r'$f(x) = 0$ para $0 x \leq x \leq 2$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');    #ejes
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-3.0,3.01]),ylim([-1.05,3.01]),grid(),legend()     # xlim y ylim plano
show()
```

Observamos que tenemos nuestra funciòn a trozos normal con nuestro paràmetro t=0




```python
# Serie de Fourier
def u_0(x,N=40): # N número de armónicos
    f = zeros(len(x))
    for n in range(1,N):   # n inicializado en 0, empieza en 1 hasta 20
        f += ((2-2*(-1)**n)/(np.pi**2 * n**2) * np.cos((n*np.pi*x)/2) + (2/(np.pi*n)) * np.sin((n*np.pi*x)/2)) # += ES UNA SUMATORIA, este f es la serie de Fourier
    return (1/2)+f  # coeficiente de a0
```

```python
x = np.concatenate((x1, x2))
f_n = u_0(x)
plot(x1, y1, label='$f(x) = 0$ para $-2 \leq x \leq 0$') # Grafica de (-2,0)
plot(x2, y2, label='$f(x) = 2-x$ para $0 x \leq x \leq 2$')   # Grafica de (0,2)
plot(x, f_n, label='$f_N(x)$ con $N=20$')                 # Grafica a trozos
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Serie de Fourier ")
xlim([-2.0,2.1]),ylim([-0.1,3.2]),grid(),legend()
show()
```

Observamos nuestra serie de fourier (aproximaciòn a nuestra funciòn original)

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.2,3.01]),ylim([-0.5,3.2]),grid(),legend()
show()
```

En esta gràfica observamos que se cumple el teorema de convergencia en una funciòn a trozos, en nuestra funciòn por la derecha converge a 2, por la izquierda converge a 1 y el promedio es 1 que es en donde nuestras funciones se intersectan
