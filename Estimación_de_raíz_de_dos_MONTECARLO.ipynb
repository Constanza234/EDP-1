RAÍZ DE 2

```python
from random import random
```

```python
import matplotlib.pyplot as plt
lx=[] #lista para mis abcisas
ly=[]
```

```python
N=1000000 #numero total de numeros generados
m=0 #exitos que cumole la desigualdad
```

```python
for i in range(N):
  x=1+random()
  if x**2<2:
    m=m+1
  if i>0 and i%1000==0:
    lx.append(i+1)
    ly.append(1+m/(i+1))
print(1+m/N)
```

```python
plt.figure(figsize=(4,2))
plt.plot(lx,ly)
plt.show()
```

```python
from random import random
```

```python
import matplotlib.pyplot as plt
lx=[]
ly=[]
```

```python
N=1000000
n=0
```

```python
for i in range (N):
  x=1+random()
  if x**2<3:
    n=n+1
  if i>0 and i%1000==0:
    lx.append(i+1)
    ly.append(1+n/(i+1))
print(1+n/N)
```

```python
plt.figure(figsize=(4,2))
plt.plot(lx,ly)
plt.show()
```
